- Приймає в конструкторі список чисел
- Має метод filter(), який фільтрує та повертає лише непарні числа
- Має property odd_numbers, яка повертає відфільтрований список непарних чисел

```python
class ProductCalculator:
    def __init__(self, numbers):
        self.numbers = numbers

    def calculate_product(self):
        result = 1
        for number in self.numbers:
            result *= number
        return result

    @property
    def product(self):
        return self.calculate_product()

    def show_student_info(self):
        print("Ім'я студента: Олександр Звізло")
        print("Група: КНк-210")


# Списки чисел для створення об'єктів
data_lists = [
    [1, 2, 3, 4],
    [2, 5, 6],
    [3, 3, 3]
]

# Створення об'єктів у циклі
calculators = []

for data in data_lists:
    calculator = ProductCalculator(data)
    calculators.append(calculator)

# Виведення результатів
for index, calculator in enumerate(calculators, start=1):
    print(f"\nОб'єкт №{index}")
    print(f"Список чисел: {calculator.numbers}")
    print(f"Добуток чисел: {calculator.product}")
    calculator.show_student_info()
```
