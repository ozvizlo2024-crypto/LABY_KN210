## –ó–∞–≤–¥–∞–Ω–Ω—è 3. –ú–µ—Ç–æ–¥–∏ –∫–ª–∞—Å—É ##

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"–í –º–µ–Ω–µ —î —Ç–≤–∞—Ä–∏–Ω–∫–∞ {animals} —è–∫—É –∑–≤–∞—Ç–∏ {name} —è–∫—ñ {age} —Ä–æ–∫—ñ–≤")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass==None else mass
    
    def speak(self):
        if self.animals == "–ö—ñ—Ç":
            return f"–ú–æ—è –¢–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –≥–æ–≤–æ—Ä–∏—Ç—å –ú—è—É—É—É"
        elif self.animals == "–°–æ–±–∞–∫–∞":
            return f"–ú–æ—è –¢–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –≥–æ–≤–æ—Ä–∏—Ç—å –ì–∞–≤–≤"
        else:
            return f"–ú–æ—è –¢–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –Ω–µ —Ä–æ–∑–º–æ–≤–ª—è—î"
    
    def add_one_year(self):
        self.age += 1
        return f"–ú–æ—è –¢–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –ø—Ä–æ–∂–∏–ª–∞ –∑–∞ –º–Ω–æ—é –æ–¥–∏–Ω —Ä—ñ–∫"

a1 = MyAnimals("–†–∏–±–∫–∏", "–ù–µ–æ–Ω–∏", 1)
a2 = MyAnimals("–ö—ñ—Ç", "–ú—É—Ä–∫–∞", 5, 10)
a3 = MyAnimals(animals="–°–æ–±–∞–∫–∞", name="–†–µ–∫—Å", age=2)

print(f"–¢–≤–∞—Ä–∏–Ω–∫–∞ {a1.animals}:", a1.speak())

print(f"–¢–≤–∞—Ä–∏–Ω–∫–∞ {a3.animals}:", a3.speak())


print(f"{a1.animals} –∂–∏–≤–µ –∑—ñ –º–Ω–æ—é {a1.age} —Ä–æ–∫—ñ–≤")
print(a1.add_one_year())
print(f"{a1.animals} –∂–∏–≤–µ –∑—ñ –º–Ω–æ—é {a1.age} —Ä–æ–∫—ñ–≤")

print(a1.age)
```

–ó–∞—Å—Ç–æ—Å—É–≤–∞—Ç–∏ –º–µ—Ç–æ–¥–∏ –∫–ª–∞—Å—É –¥–æ –≤—Å—ñ—Ö –æ–±'—î–∫—Ç—ñ–≤ –∞1-–∞3

```python
class MyAnimals:
    def __init__(self, animals, name, age, mass=None):
        print(f"–í –º–µ–Ω–µ —î —Ç–≤–∞—Ä–∏–Ω–∫–∞ {animals} —è–∫—É –∑–≤–∞—Ç–∏ {name} —è–∫—ñ {age} —Ä–æ–∫–∏")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass

    def speak(self):
        if self.animals == "–ö—ñ—Ç":
            return f"–ú–æ—è —Ç–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –≥–æ–≤–æ—Ä–∏—Ç—å –ú—è—É"
        elif self.animals == "–°–æ–±–∞–∫–∞":
            return f"–ú–æ—è —Ç–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –≥–æ–≤–æ—Ä–∏—Ç—å –ì–∞–≤"
        else:
            return f"–ú–æ—è —Ç–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –Ω–µ —Ä–æ–∑–º–æ–≤–ª—è—î"

    def add_one_year(self):
        self.age += 1
        return f"–ú–æ—è —Ç–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –ø—Ä–æ–∂–∏–ª–∞ –∑ –º–Ω–æ—é –æ–¥–∏–Ω —Ä—ñ–∫"


# —Å—Ç–≤–æ—Ä—é—î–º–æ –æ–±‚Äô—î–∫—Ç–∏
a1 = MyAnimals("–†–∏–±–∫–∏", "–ù–µ–º–æ", 1)
a2 = MyAnimals("–ö—ñ—Ç", "–ú—É—Ä–∫–∞", 5)
a3 = MyAnimals("–°–æ–±–∞–∫–∞", "–†–µ–∫—Å", 2)

# –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü—ñ—è speak()
print(f"{a1.animals}: {a1.speak()}")
print(f"{a2.animals}: {a2.speak()}")
print(f"{a3.animals}: {a3.speak()}")

# –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü—ñ—è add_one_year()
print(f"{a1.name} –º–∞—î {a1.age} —Ä–æ–∫—ñ–≤")
print(a1.add_one_year())
print(f"–¢–µ–ø–µ—Ä {a1.name} –º–∞—î {a1.age} —Ä–æ–∫—ñ–≤")
```

## –ó–∞–≤–¥–∞–Ω–Ω—è 4. –ó–º—ñ–Ω–Ω—ñ –∫–ª–∞—Å—É, –≥–ª–æ–±–∞–ª—å–Ω—ñ –∑–º—ñ–Ω–Ω—ñ –Ω–∞–¥ –≤—Å—ñ–º–∞ –æ–±'—î–∫—Ç–∞–º–∏ ##

```python
class MyAnimals:
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass = None) -> None:
        print(f"–í –º–µ–Ω–µ —î —Ç–≤–∞—Ä–∏–Ω–∫–∞ {animals} —è–∫—É –∑–≤–∞—Ç–∏ {name} —è–∫—ñ {age} —Ä–æ–∫—ñ–≤")
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass==None else mass
        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1

print(f"–í –º–µ–Ω–µ –∑–∞—Ä–∞–∑ —î —Ç–≤–∞—Ä–∏–Ω–∫–∏ {list(MyAnimals.total_animal_types)}, —ó—Ö –∫—ñ–ª—å–∫—ñ—Å—Ç—å {MyAnimals.total_animals}")
a1 = MyAnimals("–†–∏–±–∫–∏", "–ù–µ–æ–Ω–∏", 1)
MyAnimals("–†–∏–±–∫–∏", "–ì—É–ø—ñ—ó", 1)
MyAnimals("–†–∏–±–∫–∏", "–°–æ–º—ñ–∫–∏", 1)
MyAnimals("–ö—ñ—Ç", "–ú—É—Ä–∫–∞", 5, 10)
print(f"–í –º–µ–Ω–µ –∑–∞—Ä–∞–∑ —î —Ç–≤–∞—Ä–∏–Ω–∫–∏ {list(MyAnimals.total_animal_types)}, —ó—Ö –∫—ñ–ª—å–∫—ñ—Å—Ç—å {MyAnimals.total_animals}")

print(f"–Ø –¥–æ—Å—Ç—É–ø–∞—é—Å—å –¥–æ –∑–º—ñ–Ω–Ω–æ—ó –∫–ª–∞—Å—É –∑ –æ–±—î–∫—Ç–∞ {a1.total_animal_types}")
a1.total_animal_types = {"–°–æ–±–∞–∫–∞"} # —è –∑—Ä–æ–±–∏–≤ –∑–º—ñ–Ω–Ω—É –∫–ª–∞—Å—É –∞—Ç—Ä–∏–±—É—Ç–æ–º –æ–±—î–∫—Ç—É
print(f"–Ø –¥–æ—Å—Ç—É–ø–∞—é—Å—å –¥–æ –∑–º—ñ–Ω–Ω–æ—ó –∫–ª–∞—Å—É –∑ –æ–±—î–∫—Ç–∞ {a1.total_animal_types} —Ç–∞ –∑ –∫–ª–∞—Å—É {MyAnimals.total_animal_types}")
```

–ó–∞–¥–∞—Ç–∏ –¥–µ–∫—ñ–ª—å–∫–∞ –æ–±'—î–∫—Ç—ñ–≤ –≤ –∑–º—ñ–Ω–Ω—ñ–π —Ç–∞–ø –≤–∏–≤–µ—Å—Ç–∏ —ó—Ö –≥–ª–æ–±–∞–ª—å–Ω—É –∑–º—ñ–Ω–Ω—É

```python
class MyAnimals:
    # üîπ –∑–º—ñ–Ω–Ω—ñ –∫–ª–∞—Å—É (—Å–ø—ñ–ª—å–Ω—ñ –¥–ª—è –≤—Å—ñ—Ö –æ–±‚Äô—î–∫—Ç—ñ–≤)
    total_animals = 0
    total_animal_types = set()

    def __init__(self, animals, name, age, mass=None) -> None:
        print(f"–í –º–µ–Ω–µ —î —Ç–≤–∞—Ä–∏–Ω–∫–∞ {animals} —è–∫—É –∑–≤–∞—Ç–∏ {name} —è–∫—ñ {age} —Ä–æ–∫—ñ–≤")

        # üîπ –∑–≤–∏—á–∞–π–Ω—ñ –∑–º—ñ–Ω–Ω—ñ –æ–±‚Äô—î–∫—Ç–∞
        self.animals = animals
        self.name = name
        self.age = age
        self.mass = 0 if mass is None else mass

        # üîπ –æ–Ω–æ–≤–ª–µ–Ω–Ω—è –∑–º—ñ–Ω–Ω–∏—Ö –∫–ª–∞—Å—É
        MyAnimals.total_animal_types.add(animals)
        MyAnimals.total_animals += 1


# üîπ –ø–µ—Ä–µ–≤—ñ—Ä—è—î–º–æ –ø–æ—á–∞—Ç–∫–æ–≤—ñ –∑–Ω–∞—á–µ–Ω–Ω—è (–¥–æ —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –æ–±‚Äô—î–∫—Ç—ñ–≤)
print(f"–í –º–µ–Ω–µ –∑–∞—Ä–∞–∑ —î —Ç–≤–∞—Ä–∏–Ω–∫–∏ {list(MyAnimals.total_animal_types)}, "
      f"—ó—Ö –∫—ñ–ª—å–∫—ñ—Å—Ç—å {MyAnimals.total_animals}")

# üîπ —Å—Ç–≤–æ—Ä—é—î–º–æ –æ–±‚Äô—î–∫—Ç–∏ –≤ —Ü–∏–∫–ª—ñ
data = [
    ("–†–∏–±–∫–∏", "–ù–µ–æ–Ω–∏", 1),
    ("–†–∏–±–∫–∏", "–ì—É–ø—ñ—ó", 1),
    ("–†–∏–±–∫–∏", "–°–æ–º—ñ–∫–∏", 1),
    ("–ö—ñ—Ç", "–ú—É—Ä–∫–∞", 5)
]

objects = []  # —Å–ø–∏—Å–æ–∫ —Å—Ç–≤–æ—Ä–µ–Ω–∏—Ö –æ–±‚Äô—î–∫—Ç—ñ–≤

for animal, name, age in data:
    obj = MyAnimals(animal, name, age)
    objects.append(obj)

# üîπ —Å—Ç–∞–Ω –≥–ª–æ–±–∞–ª—å–Ω–∏—Ö –∑–º—ñ–Ω–Ω–∏—Ö –ø—ñ—Å–ª—è —Ü–∏–∫–ª—É
print(f"–í –º–µ–Ω–µ –∑–∞—Ä–∞–∑ —î —Ç–≤–∞—Ä–∏–Ω–∫–∏ {list(MyAnimals.total_animal_types)}, "
      f"—ó—Ö –∫—ñ–ª—å–∫—ñ—Å—Ç—å {MyAnimals.total_animals}")

# üîπ –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü—ñ—è –¥–æ—Å—Ç—É–ø—É –¥–æ –∑–º—ñ–Ω–Ω–æ—ó –∫–ª–∞—Å—É —á–µ—Ä–µ–∑ –æ–±‚Äô—î–∫—Ç
a1 = objects[0]
print(f"–Ø –¥–æ—Å—Ç—É–ø–∞—é—Å—å –¥–æ –∑–º—ñ–Ω–Ω–æ—ó –∫–ª–∞—Å—É –∑ –æ–±‚Äô—î–∫—Ç–∞ ‚Üí {a1.total_animal_types}")

# üîπ —Å—Ç–≤–æ—Ä—é—î–º–æ –∑–º—ñ–Ω–Ω—É –ª–∏—à–µ –≤ –æ–¥–Ω–æ–≥–æ –æ–±‚Äô—î–∫—Ç–∞, –∑–º—ñ–Ω—é—é—á–∏ –∑–º—ñ–Ω–Ω—É –∫–ª–∞—Å—É ‚Üí –Ω–∞ –∑–º—ñ–Ω–Ω—É –æ–±‚Äô—î–∫—Ç–∞
a1.total_animal_types = {"–°–æ–±–∞–∫–∞"}   # —Ç–µ–ø–µ—Ä —Ü–µ –í–õ–ê–°–ù–ò–ô –∞—Ç—Ä–∏–±—É—Ç a1

print(f"–¢–µ–ø–µ—Ä total_animal_types —É a1 ‚Üí {a1.total_animal_types}")
print(f"–ê —É —Å–∞–º–æ–≥–æ –∫–ª–∞—Å—É –≤–æ–Ω–∞ ‚Üí {MyAnimals.total_animal_types}")
```

## –ó–∞–≤–¥–∞–Ω–Ω—è 5. –°—Ç–∞—Ç–∏—á–Ω—ñ –º–µ—Ç–æ–¥–∏ —Ç–∞ –∑–º—ñ–Ω–Ω–∞ self ## 

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
    
    def add_one_year(self):
        self.age += 1
        return f"–ú–æ—è –¢–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –ø—Ä–æ–∂–∏–ª–∞ –∑–∞ –º–Ω–æ—é –æ–¥–∏–Ω —Ä—ñ–∫"

    @staticmethod
    def call_pet():
        return "–¢–≤–∞—Ä–∏–Ω–∫–∞ –±—ñ–∂–∏—Ç—å –¥–æ –Ω–∞—Å"


a1 = MyAnimals("–†–∏–±–∫–∏", "–ì—É–ø—ñ—ó", 1)
a2 = MyAnimals("–ö—ñ—Ç", "–ú—É—Ä–∫–∞", 5)

print(f"–ú–∏ –∫–ª–∏—á–µ–º–æ –Ω–∞—à—É —Ç–≤–∞—Ä–∏–Ω–∫—É –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Å—Ç–∞—Ç–∏—á–Ω–æ–≥–æ –º–µ—Ç–æ–¥—É –∑ –æ–±—î–∫—Ç—É |{a1.call_pet()}| —Ç–∞ –∑ –∫–ª–∞—Å—É |{MyAnimals.call_pet()}|")

print(f"–í–∏–∫–ª–∏–∫–∞—é –º–µ—Ç–æ–¥ –∑ –æ–±—î–∫—Ç—É |{a1.add_one_year()}| —Ç–∞ –∑ –∫–ª–∞—Å—É |{MyAnimals.add_one_year(a1)}|")
```

–í–∏–∫–ª–∏–∫–∞—Ç–∏ —Å—Ç–∞—Ç–∏—á–Ω–∏–π –º–µ—Ç–æ–¥ –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ –æ–±'—î–∫—Ç–∞ –∞1-–∞2

```python
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def add_one_year(self):
        self.age += 1
        return f"–ú–æ—è —Ç–≤–∞—Ä–∏–Ω–∫–∞ {self.animals} –ø—Ä–æ–∂–∏–ª–∞ –∑—ñ –º–Ω–æ—é –æ–¥–∏–Ω —Ä—ñ–∫"

    @staticmethod
    def call_pet():
        return "–¢–≤–∞—Ä–∏–Ω–∫–∞ –±—ñ–∂–∏—Ç—å –¥–æ –Ω–∞—Å"


a1 = MyAnimals("–†–∏–±–∫–∏", "–ì—É–ø—ñ–∫", 1)
a2 = MyAnimals("–ö—ñ—Ç", "–ú—É—Ä–∫–∞", 5)

# ‚úî –í–∏–∫–ª–∏–∫ —Å—Ç–∞—Ç–∏—á–Ω–æ–≥–æ –º–µ—Ç–æ–¥—É
print(f"–ö–ª–∏—á–µ–º–æ —Ç–≤–∞—Ä–∏–Ω–∫—É —Å—Ç–∞—Ç–∏—á–Ω–∏–º –º–µ—Ç–æ–¥–æ–º —ñ–∑ –æ–±‚Äô—î–∫—Ç–∞: {a1.call_pet()}")
print(f"–ö–ª–∏—á–µ–º–æ —Å—Ç–∞—Ç–∏—á–Ω–∏–π –º–µ—Ç–æ–¥ —ñ–∑ –∫–ª–∞—Å—É: {MyAnimals.call_pet()}")

# ‚úî –í–∏–∫–ª–∏–∫ add_one_year() —á–µ—Ä–µ–∑ —Ä—ñ–∑–Ω—ñ —Å–ø–æ—Å–æ–±–∏
print(f"–ú–µ—Ç–æ–¥ –≤–∏–∫–ª–∏–∫–∞–Ω–æ –∑ –æ–±‚Äô—î–∫—Ç–∞: {a1.add_one_year()}")
print(f"–ú–µ—Ç–æ–¥ –≤–∏–∫–ª–∏–∫–∞–Ω–æ –∑ –∫–ª–∞—Å—É: {MyAnimals.add_one_year(a2)}")

print(f"–ù–æ–≤–∏–π –≤—ñ–∫ a1: {a1.age}")
print(f"–ù–æ–≤–∏–π –≤—ñ–∫ a2: {a2.age}")
```

## –ó–∞–≤–¥–∞–Ω–Ω—è 7. –ú–µ—Ç–æ–¥–∏ –∫–ª–∞—Å—É ## 

```python
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data:str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)

a1 = MyAnimals.from_csv("–ö—ñ—Ç,–ú—É—Ä–∫–∞,5")
a2 = MyAnimals("–°–æ–±–∞–∫–∞", "–†–µ–∫—Å", 2)

print(f"–ú—ñ–π {a1.animals}, –∑–≤–∞—Ç–∏ {a1.name}, –º–∞—î {a1.age} —Ä–æ–∫—ñ–≤")
print(f"–ú—ñ–π {a2.animals}, –∑–≤–∞—Ç–∏ {a2.name}, –º–∞—î {a2.age} —Ä–æ–∫—ñ–≤")
print(f"–ó–∞–≥–∞–ª—å–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –æ–±—î–∫—Ç—ñ–≤ {MyAnimals.total_objects} –∞ —Å—Ç–≤–æ—Ä–µ–Ω–∏—Ö –∑ csv —Ñ–æ—Ä–º–∞—Ç—É {MyAnimals.total_csv}")
```

```python
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data: str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, int(age))

    # –ù–æ–≤–∏–π –º–µ—Ç–æ–¥ –∫–ª–∞—Å—É
    @classmethod
    def check_name(cls, obj):
        """–ü–µ—Ä–µ–≤—ñ—Ä—è—î, —á–∏ —ñ–º‚Äô—è –ø–æ—á–∏–Ω–∞—î—Ç—å—Å—è –∑ –≤–µ–ª–∏–∫–æ—ó –ª—ñ—Ç–µ—Ä–∏"""
        return obj.name[0].isupper()


# —Å—Ç–≤–æ—Ä—é—î–º–æ –æ–±‚Äô—î–∫—Ç–∏
a1 = MyAnimals.from_csv("–ö—ñ—Ç,–ú—É—Ä–∫–∞,5")
a2 = MyAnimals("–°–æ–±–∞–∫–∞", "—Ä–µ–∫—Å", 2)


# –≤–∏–≤–µ–¥–µ–Ω–Ω—è —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—ó
print(f"–ú—ñ–π {a1.animals}, –∑–≤–∞—Ç–∏ {a1.name}, –º–∞—î {a1.age} —Ä–æ–∫—ñ–≤")
print(f"–ú—ñ–π {a2.animals}, –∑–≤–∞—Ç–∏ {a2.name}, –º–∞—î {a2.age} —Ä–æ–∫—ñ–≤")

print()

# –ø–µ—Ä–µ–≤—ñ—Ä–∫–∞ –º–µ—Ç–æ–¥—É –∫–ª–∞—Å—É
print(f"–ß–∏ —ñ–º‚Äô—è a1 –ø–æ—á–∏–Ω–∞—î—Ç—å—Å—è –∑ –≤–µ–ª–∏–∫–æ—ó –ª—ñ—Ç–µ—Ä–∏? ‚Üí {MyAnimals.check_name(a1)}")
print(f"–ß–∏ —ñ–º‚Äô—è a2 –ø–æ—á–∏–Ω–∞—î—Ç—å—Å—è –∑ –≤–µ–ª–∏–∫–æ—ó –ª—ñ—Ç–µ—Ä–∏? ‚Üí {MyAnimals.check_name(a2)}")

print()

# —ñ–Ω—Ñ–æ—Ä–º–∞—Ü—ñ—è –ø—Ä–æ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –æ–±‚Äô—î–∫—Ç—ñ–≤
print(
    f"–ó–∞–≥–∞–ª—å–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –æ–±‚Äô—î–∫—Ç—ñ–≤ {MyAnimals.total_objects} "
    f"–∞ —Å—Ç–≤–æ—Ä–µ–Ω–∏—Ö –∑ CSV —Ñ–æ—Ä–º–∞—Ç—É {MyAnimals.total_csv}"
)
```
