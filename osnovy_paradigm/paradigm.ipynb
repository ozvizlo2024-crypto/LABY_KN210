# Завдання 1: Інкапсуляція

```python
import random

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner  # публічний атрибут
        self.__balance = balance  # приватний атрибут

    def deposit(self, amount):
        self.__balance += amount
        print(f"Депозит: +{amount} грн")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Зняття: -{amount} грн")
            return amount
        else:
            print(f"Спроба зняти {amount} грн: Недостатньо коштів!")
            return "Insufficient funds"

    def get_balance(self):
        return self.__balance

account = BankAccount("Богдан", 1000)

print(f"Початковий баланс {account.owner}: {account.get_balance()} грн")
print("-" * 30)

for i in range(5):
    action = random.choice(['deposit', 'withdraw'])
    amount = random.randint(100, 1000)  # Випадкова сума від 100 до 1000
    
    if action == 'deposit':
        account.deposit(amount)
    else:
        account.withdraw(amount)

print("-" * 30)
print(f"Кінцевий баланс: {account.get_balance()} грн")
```

# Завдання 2: Наслідування

```python
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        return f"{self.brand} {self.model}"

    def start_engine(self):
        return f"Двигун автомобіля {self.brand} {self.model} заведено! Врум Врум!"

class Car(Vehicle):
    def __init__(self, brand, model, seats):
        super().__init__(brand, model)
        self.seats = seats

    def display_info(self):
        return f"{super().display_info()}, Seats: {self.seats}"

car = Car("Toyota", "Camry", 5)

print(car.display_info())
print(car.start_engine())
```

# Завдання 3: Поліморфізм

```python
class Animal:
    def speak(self):
        return "..." 

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

class Fish(Animal):
    pass

animals = [Dog(), Cat(), Fish()]

for animal in animals:
    print(f"{type(animal).__name__} каже: {animal.speak()}")
```

### Відповідь до питання: 
Якщо у дочірнього класу немає власного методу, Python шукає його вгору по ланцюжку успадкування. Якщо метод відсутній і в базовому класі, виникне помилка AttributeError.

# Завдання 4: Абстракція 

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

circle = Circle(5)
print(circle.area())
```

#  Завдання 5: Гра 

```python
from abc import ABC, abstractmethod
from random import randint, choice

class Item(ABC):
    def __init__(self, name: str, health=500):
        self.name = name
        self.health = health
    
    @abstractmethod
    def attack(self, another_item):
        pass

class Sword(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self._sharp = 0
    
    def attack(self, another_item: Item):
        current_attack = self.__attack_power + self._sharp + randint(0, 10)
        another_item.health -= current_attack
        return f"Меч {self.name} завдає {current_attack} шкоди. Здоров'я {another_item.name}: {another_item.health}"
    
    def sharpening(self):
        self._sharp += 5
        return f"{self.name} заточено. Бонус до атаки підвищено."

class Axe(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power
    
    def attack(self, another_item: Item):
        current_attack = self.__attack_power + randint(0, 20)
        another_item.health -= current_attack
        return f"Сокира {self.name} завдає {current_attack} шкоди. Здоров'я {another_item.name}: {another_item.health}"

    def war_cry(self):
        self.health += 25
        return f"{self.name} видає бойовий клич та відновлює здоров'я."

class Bow(Item):
    def __init__(self, name, attack_power: int, range_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self.range_power = range_power
    
    def attack(self, another_item: Item):
        current_attack = self.__attack_power + randint(5, 15) + self.range_power
        another_item.health -= current_attack
        return f"Лук {self.name} стріляє на {current_attack} шкоди. Здоров'я {another_item.name}: {another_item.health}"
    
    def reload(self):
        self.range_power += 1
        return f"{self.name} перезаряджено. Дальність стрільби зросла."

available_weapons = [
    Sword("Ескалібур", 35),
    Axe("Руйнівник", 40),
    Bow("Срібна стріла", 25, 10)
]

player = choice(available_weapons)
enemy = choice([w for w in available_weapons if w != player])

print(f"Ваша зброя: {player.name} ({type(player).__name__})")
print(f"Ваш ворог: {enemy.name} ({type(enemy).__name__})")

while player.health > 0 and enemy.health > 0:
    print("\nВаш хід:")
    print("1. Атакувати")
    print("2. Використати підсилення")
    
    user_choice = input("Введіть номер дії: ")
    
    if user_choice == "1":
        print(player.attack(enemy))
    else:
        if isinstance(player, Sword):
            print(player.sharpening())
        elif isinstance(player, Axe):
            print(player.war_cry())
        elif isinstance(player, Bow):
            print(player.reload())
            
    if enemy.health <= 0:
        print(f"Перемога! {enemy.name} подолано.")
        break
        
    print("\nХід ворога:")
    enemy_action = randint(1, 2)
    if enemy_action == 1:
        print(enemy.attack(player))
    else:
        if isinstance(enemy, Sword):
            print(enemy.sharpening())
        elif isinstance(enemy, Axe):
            print(enemy.war_cry())
        elif isinstance(enemy, Bow):
            print(enemy.reload())
            
    if player.health <= 0:
        print(f"Поразка! {player.name} знищено.")
        break
```
