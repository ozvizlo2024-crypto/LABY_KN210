5. Створити клас OddNumbersFilter, який:

- Приймає в конструкторі список чисел
- Має метод filter(), який фільтрує та повертає лише непарні числа
- Має property odd_numbers, яка повертає відфільтрований список непарних чисел

```python
class OddNumbersFilter:
    def __init__(self, numbers):
        self.numbers = numbers
        self._odd_numbers = []

    def student_info(self):
        print("Ім'я студента: Олександр Звізло")
        print("Група: КН-210")

    def filter(self):
        self._odd_numbers = [num for num in self.numbers if num % 2 != 0]
        print(f"Відфільтровані непарні числа зі списку {self.numbers}: {self._odd_numbers}")
        return self._odd_numbers

    @property
    def odd_numbers(self):
        return self._odd_numbers


# Списки чисел для створення об'єктів
lists_of_numbers = [
    [1, 2, 3, 4, 5],
    [10, 11, 12, 13, 14],
    [7, 8, 9, 10]
]

# Створення об'єктів у циклі
objects = []

for numbers in lists_of_numbers:
    obj = OddNumbersFilter(numbers)
    objects.append(obj)

# Робота з кожним об'єктом
for obj in objects:
    obj.student_info()
    obj.filter()
    print(f"Властивість odd_numbers повертає: {obj.odd_numbers}")
    print("-" * 50)
```
