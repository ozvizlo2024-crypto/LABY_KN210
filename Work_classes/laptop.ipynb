## Виконати приклади які розглядали на лекції та вставити їх у Ноутбук

```python
import random

class Fighter:
    def __init__(self, title):
        self._title = title
        self.hp = 100
        self.damage_value = 20

    def strike(self, enemy):
        hit = BattleMath.get_hit(self.damage_value)
        enemy.hp -= hit
        return f"{self._title} наносить удар по '{enemy._title}'! HP сперника: {enemy.hp}"

    def restore(self):
        heal_points = random.randint(5, 15)
        self.hp += heal_points
        return f"{self._title} відновлює {heal_points} HP. Поточний стан: {self.hp}"

class BattleMath:
    @staticmethod
    def get_hit(amount):
        chance = random.random()
        if chance < 0.1:
            print("⚡ Суперудар! Подвійний збиток!")
            return amount * 2
        elif chance < 0.2:
            print("❌ Удар не попав!")
            return 0
        return amount

    @staticmethod
    def show_round(num):
        print(f"\n=== Раунд №{num} ===")

player = Fighter("Герой")
enemy = Fighter("Ворог")

round_counter = 1

while round_counter <= 3:
    BattleMath.show_round(round_counter)

    # Хід гравця
    if random.randint(0, 1):
        print(player.strike(enemy))
    else:
        print(player.restore())

    # Хід монстра
    if random.randint(0, 1):
        print(enemy.strike(player))
    else:
        print(enemy.restore())

    round_counter += 1
```
